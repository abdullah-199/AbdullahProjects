```python
import os
from datetime import datetime

# Define a class for books
class Book:
    def __init__(self, isbn, name, author, edition, publisher, yop, status="available"):
        self.isbn = isbn
        self.name = name
        self.author = author
        self.edition = edition
        self.publisher = publisher
        self.yop = yop
        self.status = status
        self.borrowed_by = None
        self.borrowed_date = None

# Define a class for students
class Student:
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll

# List to store books and issued books
books = []
issued_books = []

# Function to add a book
def add_book():
    isbn = input("Enter book ISBN number: ")
    name = input("Enter book name: ")
    author = input("Enter author's name: ")
    edition = input("Enter book edition: ")
    publisher = input("Enter book publisher: ")
    yop = input("Enter year of publication: ")
    
    book = Book(isbn, name, author, edition, publisher, yop)
    books.append(book)
    print("Book added successfully!")

# Function to list all books
def list_books():
    print("\nAvailable Books:")
    for i, book in enumerate(books):
        print(f"{i+1}. {book.name} by {book.author} ({book.status})")

# Function to remove a book
def remove_book():
    isbn = input("Enter book ISBN to remove: ")
    for book in books:
        if book.isbn == isbn:
            books.remove(book)
            print("Book removed successfully!")
            return
    print("Book not found!")

# Function to issue a book
def issue_book():
    isbn = input("Enter book ISBN to issue: ")
    for book in books:
        if book.isbn == isbn and book.status == "available":
            student_name = input("Enter student name: ")
            student_roll = input("Enter student roll number: ")
            book.status = "borrowed"
            book.borrowed_by = Student(student_name, student_roll)
            book.borrowed_date = datetime.now().strftime("%Y-%m-%d")
            issued_books.append(book)
            print(f"Book '{book.name}' issued to {student_name}.")
            return
    print("Book not available or not found!")

# Function to list issued books
def list_issued_books():
    print("\nIssued Books:")
    for i, book in enumerate(issued_books):
        print(f"{i+1}. {book.name} by {book.author} (Issued to: {book.borrowed_by.name}, Roll: {book.borrowed_by.roll}, Date: {book.borrowed_date})")

# Function to return a book
def return_book():
    isbn = input("Enter book ISBN to return: ")
    for book in issued_books:
        if book.isbn == isbn:
            book.status = "available"
            book.borrowed_by = None
            book.borrowed_date = None
            issued_books.remove(book)
            print("Book returned successfully!")
            return
    print("Book not found in issued list!")

# Main function to run the library management system
def main():
    while True:
        print("\nLibrary Management System")
        print("1. Add Book")
        print("2. List Books")
        print("3. Remove Book")
        print("4. Issue Book")
        print("5. List Issued Books")
        print("6. Return Book")
        print("0. Exit")
        
        choice = input("Enter your choice: ")
        
        if choice == "1":
            add_book()
        elif choice == "2":
            list_books()
        elif choice == "3":
            remove_book()
        elif choice == "4":
            issue_book()
        elif choice == "5":
            list_issued_books()
        elif choice == "6":
            return_book()
        elif choice == "0":
            print("Exiting...")
            break
        else:
            print("Invalid choice! Please try again.")

if __name__ == "__main__":
    main()
```
